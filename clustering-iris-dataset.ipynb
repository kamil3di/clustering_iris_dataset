## Importing libraries

```python
import numpy as np
import pandas as pd
from sklearn import datasets
```

## Loading iris data set

```python
iris = datasets.load_iris()
```

```python
iris_data = pd.DataFrame(iris.data)
iris_data.info()
```

## Model fitting and prediction

```python
from sklearn.cluster import KMeans

model = KMeans(n_clusters=3)
model.fit(iris_data)
labels1 = model.predict(iris_data)
labels1
```

```python
import matplotlib.pyplot as plt

#sepal length (cm)
xs1 = iris_data.iloc[:,0]
#sepal width (cm)
ys1 = iris_data.iloc[:,1]

centroids = model.cluster_centers_
centroids_x1 = centroids[:,0]
centroids_y1 = centroids[:,1]
```

```python
plt.scatter(xs1, ys1, c=labels1, alpha=0.5)
plt.scatter(centroids_x1, centroids_y1, marker='D', s=50)
plt.title('Sepal length(cm) vs sepal width(cm)')
plt.xlabel('Sepal length(cm)')
plt.ylabel('Sepal width(cm)')
plt.show()
```

```python
#petal length (cm)
xs2 = iris_data.iloc[:,2]
#petal width (cm)
ys2 = iris_data.iloc[:,3]

centroids_x2 = centroids[:,2]
centroids_y2 = centroids[:,3]
```

```python
plt.scatter(xs2, ys2, c=labels1, alpha=0.5)
plt.scatter(centroids_x2, centroids_y2, marker='D', s=50)
plt.title('Petal length(cm) vs petal width(cm)')
plt.xlabel('Petal length(cm)')
plt.ylabel('Petal width(cm)')
plt.show()
```

## Creating varieties

```python
# 0's are  setosa,1's are versicolor,2's are virginica
iris.target_names
```

```python
varieties = []
for target in iris.target:
    if(target==0):
        varieties.append('setosa')
    elif(target==1):
        varieties.append('versicolor')
    else:
        varieties.append('virginica')
```

## Evaluating

```python
df = pd.DataFrame({'labels': labels1, 'varieties':varieties})

ct = pd.crosstab(df['labels'], df['varieties'])

print(ct)
```
